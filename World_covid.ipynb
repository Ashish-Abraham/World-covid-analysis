```python
import pandas as pd
data=pd.read_csv('/content/files/owid-covid-data.csv')
```

```python
data.describe()
```

```python
data.tail()

```

```python
israel_covid=data.loc[40909:40994][['location','total_cases','weekly_icu_admissions','new_deaths','date']]
israel_covid
```

```python
import matplotlib.pyplot as plt
x=list(israel_covid['date'])
y=list(israel_covid['new_deaths'])
plt.figure(figsize=(10,10))
plt.scatter(x,y,c=y,s=100,cmap='rainbow',edgecolor='none')
plt.colorbar()
plt.show()
plt.plot(x,y)
```

Israel was one of the most successful nations in fighting covid. The plot clearly explains the decrease in covid deaths upto May 2021. 

```python
china_covid=data.loc[data['location']=='China'][['location','total_cases','weekly_icu_admissions','new_deaths','date']]
china_covid
```

```python
china_covid_may=china_covid.loc[17577:17600]
x=list(china_covid_may['date'])
y=list(china_covid_may['new_deaths'])
plt.figure(figsize=(10,10))
plt.scatter(x,y,c=y,s=100,cmap='inferno',edgecolor='none')
plt.colorbar()
plt.show()
plt.plot()
```

Analysis of data shows that covid deaths in China has remained 0 in the whole month of May. Some inaccuracy in data is clear here.

```python
peru_covid=data.loc[data['location']=='Peru'][['location','total_cases','weekly_icu_admissions','new_deaths','date']]
peru_covid_march_may=peru_covid.loc[65180:65264]
peru_covid_march_may
```

```python
x=list(peru_covid_march_may['date'])
y=list(peru_covid_march_may['new_deaths'])
plt.scatter(x,y,s=40,c=y,cmap='hsv',edgecolor='none')
plt.colorbar()
plt.show()
plt.plot(x,y)
```

Analysing scenario in South America, we take Peru to explain whats happening. Covid deaths has drastically increased to alarming numbers and a steep rate during January 2021 to April 2021. May 2021 is seeing better conditions and we hope for better.
